```python
movie_rec=pd.DataFrame()
movie_rec['movie']=df_corr['title']
movie_rec
```

```python
m=0
movie_rec['recommend']=''      #creating a blank column with null values in dataframe
movie_rec['recommend']=movie_rec['recommend'].astype(object)    #to get rid of sequence error the column should be object rather than float or int to have list as values
for i in list(df_corr['title']):
    l=list(corr_r.loc[:,i].sort_values(ascending=False).index)
    if movie_rec.loc[m,'movie']==l[0]:
        movie_rec.at[m,'recommend']=l[1:]     # removing first correlation as it is with itself
    else:
        movie_rec.at[m,'recommend']=l         # for <80 ratings first correlation would not be with itself so not removing 
    m+=1
```

```python
movie_rec['user_id']=df_corr['user_id']
movie_rec=movie_rec[['movie','user_id','recommend']]
movie_rec
```

# Removing already watched movies

```python
movie_rec['final_recommend']=''
movie_rec['final_recommend']=movie_rec['final_recommend'].astype(object)
for i in movie_rec['user_id']:
    movie_rec.at[i,'final_recommend']=[x for x in movie_rec.loc[i,'recommend'] if x not in list(watched.loc[i,'watched_movies'])]
```
